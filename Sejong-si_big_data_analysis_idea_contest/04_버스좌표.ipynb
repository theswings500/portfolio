```python
import pandas as pd
import requests, json

def get_address(lat, lng):
    url = "https://dapi.kakao.com/v2/local/geo/coord2regioncode.json?x="+str(lng)+"&y="+str(lat)
    headers = {"Authorization": "KakaoAK 266ad1ea2bfd32a9ddb69530dd1a57e0"}
    api_json = requests.get(url, headers=headers)
    full_address = json.loads(api_json.text)

    return full_address['documents'][0]['address_name']


df['주소'] = df.apply(lambda row: get_address(row['GPS_Y'], row['GPS_X']), axis=1)

df
```

```python
df['주소']
```

```python
df3 = df[df['주소'].str.contains('세종특별자치시')]
```

```python
df3['주소']
```

```python
df3['주소2'] = df3['주소'].str.replace('세종특별자치시 ', '')  # '세종특별자치시' 삭제
df3['주소2'] = df3['주소2'].str.replace(r'(면\s).*', r'\1', regex=True)  # '면' 이후의 부분 삭제
df3['주소2'] = df3['주소2'].str.replace(r'(읍\s).*', r'\1', regex=True)  # '면' 이후의 부분 삭제
```

```python
df3['주소2'] = df3['주소2'].str.rstrip()
```

```python
print(df3['주소2'].unique())
```

```python
df_pop = pd.read_csv('세종시 인구 데이터 완료2023.csv', encoding = 'euc-kr')
```

```python
df_pop = df_pop[['ADM_CD', '도시']]
df_pop
```

```python
df_pop = df_pop.drop_duplicates()
df_pop.head()
```

```python
df_pop['주소2'] = df_pop['도시']
df_pop = df_pop[['ADM_CD', '주소2']]
df_pop.head()
```

```python
df_merged = df3.merge(df_pop, on='주소2', how='inner')
# df_merged = df3.merge(df_pop, on='주소2', how='left')
```
