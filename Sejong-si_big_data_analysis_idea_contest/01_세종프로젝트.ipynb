# 1. 버스 정류장 좌표 데이터

```python
df = pd.read_csv('세종특별자치시_정류장정보_20230602.csv', encoding = 'euc-kr')
```

```python
df_n = pd.read_csv('세종특별자치시_노선별 정류장현황_20230602.csv', encoding = 'euc-kr')
```

```python
df.head()
```

```python
df_n.head()
```

세종특별자치시_정류장정보_20230602.csv, 세종특별자치시_노선별 정류장현황_20230602.csv 2개의 데이터프레임을 정류장ID를 기준으로 병합
* 병합한 이유 : 세종특별자치시_정류장정보_20230602.csv 파일의 정류장x위치값, 정류장y위치값의 좌표계를 특정할 수 없다.

```python
merged_df = df.merge(df_n, on='정류장ID')
merged_df
```

### 병합한 데이터프레임의 모든 열의 결측치 개수 출력

```python
print(merged_df.isna().sum())
```

### 병합한 데이터프레임의 '정류장ID'열의 고유한 값의 개수 출력

```python
print(merged_df['정류장ID'].nunique()) # 정류장ID으로 고유한 값 판별
# 데이터프레임의 전체 행의 개수는 3639개이지만, 고유한 정류장ID의 개수는 1485개
```

### '정류장ID'의 고유한 값을 제외한 나머지 행 삭제

```python
df_clean = merged_df.drop_duplicates(subset='정류장ID')
```

```python
df_clean.shape
```

```python
df_clean.head()
```

### df_clean 데이터프레임 저장

```python
df_clean.to_csv('df_clean.csv', index=False, encoding = 'euc-kr')
```

# 2. 행정동별 인구 데이터 전처리

```python
import pandas as pd

# csv파일 불러오기
df_pop = pd.read_csv('세종시_주민등록인구(2013~2022)/행정구역_읍면동_별_5세별_주민등록인구_2023년_8월_20230905103757.csv', encoding = 'euc-kr')
```

```python
df_pop.shape
```

```python
df_pop.head()
```

```python
df_pop2 = df_pop
# 첫 번째 행을 열 이름으로 설정
df_pop2.columns = df_pop2.iloc[0]
# 데이터프레임에서 첫 번째 행(헤더) 제거
df_pop2 = df_pop2.drop(index=0).reset_index(drop=True)



# 결과 출력
df_pop2.head(3)
```

```python
converted_data = [] # 변환된 데이터를 저장할 리스트 생성

# df_pop2를 순회하며 새로운 데이터 생성
for i in range(len(df_pop2)):
    for j in range(2, len(df_pop2.columns)):
        converted_data.append({"도시": df_pop2.iloc[i, 0], "인구수": df_pop2.iloc[i, j],  "나이": df_pop2.columns[j]})

# 새 데이터프레임 생성
new_df = pd.DataFrame(converted_data)
new_df
```

```python
# 중복된 행 제거
unique_df = new_df.drop_duplicates().reset_index(drop= True)
unique_df
```

```python
unique_df2 = unique_df.drop(unique_df[unique_df['도시']=='세종특별자치시'].index).reset_index(drop = True)
unique_df2
```

# 3. 법정동명 전처리

```python
df_legal = pd.read_csv('세종시 법정동명.txt', sep='|', encoding='euc-kr')
```

```python
df_legal.shape
```

```python
df_legal.head(4)
```

LOWEST_ADM_SECT_NM열에서 ~리가 포함되어있으면 drop

```python
df_legal2 = df_legal.copy()
df_legal2 = df_legal2[~df_legal2['LOWEST_ADM_SECT_NM'].str.endswith('리')]
```

```python
df_legal2.head()
```

```python
df_legal2 = df_legal2.drop(df_legal2[df_legal2['LOWEST_ADM_SECT_NM']=='세종특별자치시'].index).reset_index(drop = True)
df_legal2.head()
```

```python
df_legal2.shape
```

```python
df_legal2 = df_legal2.drop(columns=['ADM_SECT_GBN', 'ADM_SECT_NM', 'DEL_YMD', 'CHG_BEF_ADM_SECT_GBN', 'CRE_YMD', 'CHG_BEF_ADM_SECT_CD', 'COL_ADM_SECT_CD'])
df_legal2.head()
```

```python
df_legal2['ADM_CD'] = df_legal2['ADM_CD'].astype(str).str.slice(0, -2)
```
